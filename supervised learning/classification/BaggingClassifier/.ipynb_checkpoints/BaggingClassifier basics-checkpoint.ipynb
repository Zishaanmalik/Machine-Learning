# Bagging Classifier (Basic Example with Logistic Regression)

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D  # for 3D plotting
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import BaggingClassifier
from sklearn.model_selection import train_test_split
```

```python
#Create a simple dataset
# Features (3 dimensions)
X = []
y = []

for i in range(3):          # feature 1
    for j in range(3):      # feature 2
        for k in range(3):  # feature 3
            X.append([i, j, k])
            # target rule: simple sum-based threshold
            y.append(1 if (i + j + k) >= 5 else 0)
```

```python
X = np.array(X)
y = np.array(y)
```

```python
X,y
```

```python
# Train-test split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
# Bagging Classifier with Logistic Regression
base_model = LogisticRegression()
bag_model = BaggingClassifier(estimator=base_model, n_estimators=10, random_state=42)
bag_model.fit(X_train, y_train)
y_pred = bag_model.predict(X_test)
```

```python
# Results
print("Model Score (accuracy):", bag_model.score(X_test, y_test))
```

```python
# Visualization
fig = plt.figure(figsize=(12, 10))
```

```python
# Visualization
fig = plt.figure(figsize=(12, 10))

# Plot 1: Feature1 vs Feature2
ax1 = fig.add_subplot(2, 2, 1)
ax1.scatter(X_train[:, 0], X_train[:, 1], c=y_train, marker='o', label="Train")
ax1.scatter(X_test[:, 0], X_test[:, 1], c=bag_model.predict(X_test), marker='s', label="Test Pred")
ax1.set_xlabel("Feature 1")
ax1.set_ylabel("Feature 2")
ax1.set_title("Feature1 vs Feature2")
ax1.legend()

# Plot 2: Feature2 vs Feature3
ax2 = fig.add_subplot(2, 2, 2)
ax2.scatter(X_train[:, 1], X_train[:, 2], c=y_train, marker='o')
ax2.scatter(X_test[:, 1], X_test[:, 2], c=bag_model.predict(X_test), marker='s')
ax2.set_xlabel("Feature 2")
ax2.set_ylabel("Feature 3")
ax2.set_title("Feature2 vs Feature3")

# Plot 3: Feature1 vs Feature3
ax3 = fig.add_subplot(2, 2, 3)
ax3.scatter(X_train[:, 0], X_train[:, 2], c=y_train, marker='o')
ax3.scatter(X_test[:, 0], X_test[:, 2], c=bag_model.predict(X_test), marker='s')
ax3.set_xlabel("Feature 1")
ax3.set_ylabel("Feature 3")
ax3.set_title("Feature1 vs Feature3")

# Plot 4: 3D plot
ax4 = fig.add_subplot(2, 2, 4, projection='3d')
ax4.scatter(X_train[:, 0], X_train[:, 1], X_train[:, 2], c=y_train, marker='o', label="Train")
ax4.scatter(X_test[:, 0], X_test[:, 1], X_test[:, 2],
            c=bag_model.predict(X_test), marker='^', label="Test Pred")
ax4.set_xlabel("Feature 1")
ax4.set_ylabel("Feature 2")
ax4.set_zlabel("Feature 3")
ax4.set_title("3D Feature Space")
ax4.legend()

plt.tight_layout()
plt.show()
```
